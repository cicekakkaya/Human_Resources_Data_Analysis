```python
# First I find out how many people are in each department.
# I create a graph of turnover by department.
```

```python
for i in data_frame['Department'].unique():
    plt.hist(data_frame[data_frame['Department'] ==i]['Attrition'], bins=6, alpha=0.5, label=i)
plt.xlabel('Attrition')
plt.ylabel('Department')
plt.title('Departmana Göre İşten Ayrılma Grafiği')
plt.legend()
plt.show()
```

```python
'''
According to this graph, the department with the highest turnover rate is the Sales department. 
This is followed by the R&D department, 
while the department with the least turnover is the Human Resources department.
'''
```

# 4. Income Analysis:

```python
# Let's analyze the distribution of monthly income by departments. “MonthlyIncome” “Department”
```

```python
monhly_income = data_frame["MonthlyIncome"].value_counts()
monhly_income
```

```python
for i in data_frame['Department'].unique():
    plt.hist(data_frame[data_frame['Department'] == i]['MonthlyIncome'], bins=6, alpha=0.5, label=i)

plt.xlabel('MonthlyIncome')
plt.ylabel('Department')
plt.title('Departmana Göre Aylık Gelir Grafiği')
plt.legend()
plt.show()
```

```python
""" 
According to this graph, while the R&D department has the lowest monthly income, 
the majority of the departments with the highest salaries are
employees in the R&D department.

It is seen that a significant portion of the Sales department is below average in terms of monthly salary.

In the Human Resources department, it is seen that there are a majority of employees with low salaries, 
while there are not many employees from this department in the average salary scale. 
But, in the highest salary scale, it is seen that there are many employees in the HR department, 
compared to the number of employees, and even the highest number of employees compared to 3 departments.

As a result, we can say that the HR department is overpaid compared to other departments. 
Since this situation will cause income injustice, it is obvious that it may lead to turnover. 
A more balanced salary distribution may be recommended in this regard.
"""
```

```python
# Let's analyze the distribution of monthly incomes by job roles. “JobRole” “MonthlyIncome”
```

```python
for i in data_frame['JobRole'].unique():
    plt.hist(data_frame[data_frame['JobRole'] == i]['MonthlyIncome'], bins=6, alpha=0.5, label=i)

plt.xlabel('MonthlyIncome')
plt.ylabel('JobRole')
plt.title('İş Rollerine Göre Aylık Gelir Grafiği')
plt.legend()
plt.show()
```

```python
"""
According to the graph above, employees in managerial positions earn the highest salary, at 20,000. Employees in job 
Roles such as Research Scientists, Human Resources, Lab Technicians are some of the lowest paid positions.

Sales Professionals are in the middle-lower segment.

However, employees in the Research Manager job role are in the upper-middle segment.

From here, it is seen that the salaries of those working in management positions have increased at 
a much higher rate than other positions, and that there is an excessive increase in monthly salaries 
in the form of a J Curve towards managerial positions. 
We can say that there is an income injustice.
"""
```

```python
# Let's investigate whether there is a relationship between income and job satisfaction. “JobSatisfaction” “MonthlyIncome”
```

```python
plt.scatter(data_frame['JobSatisfaction'], data_frame['MonthlyIncome'])
plt.xlabel('JobSatisfaction')
plt.ylabel('MonthlyIncome')
plt.title('İş Tatmini ile Aylık Gelir Arasındaki İlişki')
plt.grid(True)
plt.show()
```

```python
"""
The graph above shows that there is no significant relationship with 
job satisfaction up to the monthly income level in the upper middle segment. 

However, it is mostly at low levels in the upper middle segment.

This shows us that, as the salary increases, the expected job satisfaction is not met by the company.

In the highest salary scale, there is again no significant relationship.
"""
```
